##Trên 1 ảnh

```python
im = Image.open("/content/gdrive/MyDrive/CS114.N11.KHCL - Máy học/Fresh and rotten fruit classification/Data/Original Image/FreshApple/FreshApple (2).jpg")
im = im.resize((200, 200))
print("im")
im.show()

# rotate image degrees
angle = 30
for i in range (0,3):
  im_ro = im.rotate(angle , expand=True)
  print(f"im_ro {angle} ")
  angle = angle + 30
  im_ro.show()

#Brightness image degrees
for i in (0.5, 1.5):
  im_Br = ImageEnhance.Brightness(im).enhance(i)
  print(f"im_Br {i} ")
  im_Br.show()

#Contrast image degrees
for i in (0.5, 1.5):
  im_Co = ImageEnhance.Contrast(im).enhance(i)
  print(f"im_Co {i} ")
  im_Co.show()

#Sharpness image
for i in (-2,5):
  im_Sh = ImageEnhance.Sharpness(im).enhance(i)
  print(f"im_Sh {i} ")
  im_Sh.show()
```

```python
folder_path = 'Split Original'
save_path = "/content/gdrive/MyDrive/CS114.N11.KHCL - Máy học/Fresh and rotten fruit classification/Data/Split PreProcessing"
file_path = [os.path.join(folder_path,d) for d in os.listdir(folder_path)]
for d in os.listdir(folder_path):
  split_paths = [os.path.join(folder_path,d)]
  for split_path in split_paths:
    file_path =  [os.path.join(split_path,d) for d in os.listdir(split_path)]
    c_image = 0
    for path in file_path:
      count = 0
      for image in tqdm.tqdm(os.listdir(path)):
        path_image_save = os.path.join(save_path,os.path.basename(split_path),os.path.basename(path),os.path.basename(path))
        c_image += 1
        im = Image.open(os.path.join(path,image))

        #rotate image 
        angle = 0
        for i in range (0,4):
          im_ro = im.rotate(angle , expand=True)
          count += 1
          im_ro.save(f"{path_image_save}" +str(count) +".jpg")
          angle = angle + 30
      
        #Brightness image
        for i in (0.5, 1.5):
          im_Br = ImageEnhance.Brightness(im).enhance(i)
          count += 1
          im_Br.save(f"{path_image_save}" +str(count) +".jpg")
        
        #Contrast image
        for i in (0.5, 1.5):
          im_Co = ImageEnhance.Contrast(im).enhance(i)
          count += 1
          im_Co.save(f"{path_image_save}" +str(count) +".jpg")
            
        #Sharpness image
        for i in (-3,5):
          im_Sh = ImageEnhance.Sharpness(im).enhance(i)
          count += 1
          im_Sh.save(f"{path_image_save}" +str(count) +".jpg")
        
        print('\t',"count:",count,'\t',image)
    print("count_image:",c_image)
```
